```python
from scipy.io import arff
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np


data = pd.read_csv('/Users/vishalnigam/Downloads/1429_1.csv')
data
```

```python
features = data[['reviews.text']]
```

```python
labels= data[['categories']]
```

```python
vect_word = CountVectorizer()
```

```python
reviews = vect_word.fit_transform(features['reviews.text'].values.astype('U'))
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test,train_labels,test_labels = train_test_split(reviews,labels,test_size = 0.33)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(train, train_labels)
preds = nb.predict(test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(test_labels,preds))
```
